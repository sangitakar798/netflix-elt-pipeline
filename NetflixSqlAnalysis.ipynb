```python
import sqlalchemy as sal

engine = sal.create_engine('mysql+pymysql://root:@localhost:3306/netflixdataanalysisdb')

conn = engine.connect()
print("Connection Successful!")
```

```python
df.to_sql('netflix_raw_bkp', con=engine, index=False, if_exists='append')

print("Done! Rows inserted:", len(df))
```
